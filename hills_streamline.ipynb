```python
import numpy as np
```

```python
xl = -2
xr = 2
yl = -2
yr = 2

a = 1

x = np.linspace(xl, xr, 500)
y = np.linspace(yl, yr, 500)

X, Y = np.meshgrid(x, y)
dist = np.sqrt(X**2 + Y**2)
psi = np.zeros_like(X)

psi[dist <= a] = Y[dist <= a]**2 *(1-Y[dist <= a]**2-X[dist <= a]**2)/10
psi[dist > a] = Y[dist > a]**2 * (1/((X[dist > a]**2+Y[dist > a]**2)**1.5) -1)/15
```

```python
import matplotlib.pyplot as plt

plt.contour(X, Y, psi, 300,colors='blue', linewidths=0.1)   
plt.xlim(xl, xr)
plt.ylim(yl, yr)
plt.colorbar()
plt.axis('equal')
```
