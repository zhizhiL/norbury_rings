```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
FrSt_thin = np.array([1, 1/3, 0.25,0.64,	4,	1,	4])
vol_den_thin = np.array([12.32065593,	7.82950065,	5.643965193	,	12.61486263,	13.17925915,	13.46746162,	11.60615395])

FrSt_fat = np.array([1,	0.25,	0.64,		4])
vol_den_fat = np.array([3.227267331,	0.082558001,	1.401984971,		4.168428548])

FrSt_hill = np.array([0.64, 1, 0.25, 4])
vol_den_hill = np.array([0.056199483,	0.110477617		,0.011528099,	1.456863526])
```

```python
reds = cm.get_cmap('Reds', 7)

plt.loglog(FrSt_thin, vol_den_thin, 'o', markersize=10, label=r'Thin ($\alpha=0.4$)',color=reds(2))
plt.loglog(FrSt_fat, vol_den_fat, 'o',markersize=10,label=r'Fat ($\alpha=0.8$)',color=reds(4))
plt.loglog(FrSt_hill, vol_den_hill, 'o', markersize=10,label=r'Spherical ($\alpha=\sqrt{2}$)',color=reds(6))
plt.loglog(np.linspace(0.2, 5, 100), np.ones(100), 'k--')
plt.legend()
plt.xlabel(r'$Fr^2/St$',fontsize=15)
plt.ylabel(r'$\frac{\phi(t=5\tau_f)}{\phi_\infty}$', rotation='horizontal',fontsize=15)
```
